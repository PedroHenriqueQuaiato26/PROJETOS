
###**Metadata Bsica**

**version:** A verso do seu projeto seguindo o versionamento semntico (semver), geralmente no formato major.minor.patch.

```python
"version": "1.0.0"
```

**main**: O ponto de entrada do seu aplicativo. Geralmente  o arquivo principal que inicia o servidor.

###**Scripts**

**scripts**: Scripts que voc pode executar com npm run O script start  convencionalmente usado para iniciar a aplicao.

```python
"scripts": {
  "start": "node server.js",
  "dev": "nodemon server.js"
}
```

**start:** Inicia o aplicativo usando Node.js.

**dev:** Inicia o aplicativo usando nodemon, que reinicia o servidor automaticamente quando arquivos so modificados. Isso  til para desenvolvimento.

**keywords:** Uma lista de palavras-chave que descrevem o projeto. Essas palavras ajudam na busca do seu projeto se ele for publicado no npm.

```python
"keywords": [
  "node",
  "express",
  "postgresql",
  "exemplo"
]
```

**author:** Informao sobre o autor do projeto.

```python
"author": "Seu Nome <seuemail@example.com>"
```

###License

**license:** A licena sob a qual o projeto  distribudo. O MIT  uma licena permissiva comum.


```python
"license": "MIT"
```

###Dependencies

**dependencies:** As dependncias do projeto que so necessrias para a execuo em produo. Quando voc executa **npm install**, estas dependncias so instaladas.

```python
"dependencies": {
  "express": "^4.17.1",
  "pg": "^8.6.0"
}
```

###DevDependencies

**devDependencies:** Dependncias que so necessrias apenas para o desenvolvimento e no para a produo. Quando voc executa npm install, estas dependncias so instaladas apenas se voc no usar a flag **--production.**

```python
"devDependencies": {
  "nodemon": "^2.0.7"
}
```

**nodemon:** Uma ferramenta que reinicia automaticamente o servidor Node.js quando detecta alteraes no cdigo.

## Estrutura do Projeto

image.png

## Configurao do Projeto

```python
mkdir calculadora-node
cd calculadora-node
npm init -y
```

## Criar o backend com Express

```python
const express = require('express');
const path = require('path');

const app = express();
const PORT = 3000;

// Servir arquivos estticos (HTML, CSS)
app.use(express.static(path.join(__dirname, 'public')));

// Endpoint para calcular
app.get('/calculate', (req, res) => {
    const { operacao, num1, num2 } = req.query;
    const numero1 = parseFloat(num1);
    const numero2 = parseFloat(num2);
    let resultado;

    switch (operacao) {
        case 'adio':
            resultado = numero1 + numero2;
            break;
        case 'subtrao':
            resultado = numero1 - numero2;
            break;
        case 'multiplicao':
            resultado = numero1 * numero2;
            break;
        case 'diviso':
            if (numero2 !== 0) {
                resultado = numero1 / numero2;
            } else {
                return res.send('Erro: diviso por zero');
            }
            break;
        default:
            return res.send('Operao invlida');
    }

    res.send(`Resultado: ${resultado}`);
});

app.listen(PORT, () => {
    console.log(`Servidor rodando em http://localhost:${PORT}`);
});
```

## Executar o script

```python
node calculadora.js
```

## Front-and do projeto em HTML E CSS

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Calculadora Web</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            display: flex;
            justify-content: center;
            align-items: center;
            height: 100vh;
            background-color: #f0f0f0;
        }
        .calculator {
            background: white;
            padding: 20px;
            border-radius: 5px;
            box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
        }
        .calculator input, .calculator select {
            margin: 10px 0;
            padding: 10px;
            font-size: 1em;
            width: 100%;
        }
        .calculator button {
            padding: 10px;
            font-size: 1em;
            width: 100%;
            background-color: #007bff;
            color: white;
            border: none;
            border-radius: 5px;
        }
        .calculator button:hover {
            background-color: #0056b3;
        }
    </style>
</head>
<body>
    <div class="calculator">
        <h1>Calculadora Web</h1>
        <select id="operacao">
            <option value="adio">Adio</option>
            <option value="subtrao">Subtrao</option>
            <option value="multiplicao">Multiplicao</option>
            <option value="diviso">Diviso</option>
        </select>
        <input type="number" id="num1" placeholder="Primeiro nmero">
        <input type="number" id="num2" placeholder="Segundo nmero">
        <button onclick="calcular()">Calcular</button>
        <p id="resultado"></p>
    </div>

    <script>
        function calcular() {
            const operacao = document.getElementById('operacao').value;
            const num1 = document.getElementById('num1').value;
            const num2 = document.getElementById('num2').value;

            fetch(`/calculate?operacao=${operacao}&num1=${num1}&num2=${num2}`)
                .then(response => response.text())
                .then(data => {
                    document.getElementById('resultado').innerText = data;
                })
                .catch(error => {
                    console.error('Erro:', error);
                });
        }
    </script>
</body>
</html>
```
